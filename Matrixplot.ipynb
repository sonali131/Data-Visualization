```python
import seaborn as sns

# Load the dataset first
tips = sns.load_dataset('tips')

# Now select the required columns
tipscrror = tips[['total_bill', 'tip', 'size']]
tipscrror.head()
```

```python
tipscrror
```

```python
tipscrror.corr()
```

```python
sns.heatmap(tipscrror.corr(),annot=True)
```

```python
#Cluster map
sns.clustermap(tipscrror.corr())
```

```python
# pivot table Heat map
flights = sns.load_dataset('flights')
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
