```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read CSV (upload first if needed)
df = pd.read_csv('matches.csv')

# Step 2: Remove rows with missing bowling data
df = df.dropna(subset=['best_bowling', 'best_bowling_figure'])

# Step 3: Extract wickets from the figure like '4--18'
df['highest_wickets'] = (
    df['best_bowling_figure']
      .astype(str)
      .str.split('--')
      .str[0]
      .astype(float)
)

# Step 4: Check which bowler took the highest wickets
top_bowler = df[df['highest_wickets'] == df['highest_wickets'].max()][
    ['best_bowling', 'best_bowling_figure', 'highest_wickets']
]

print("ðŸ Top Bowler(s) with Highest Wickets:")
print(top_bowler)

# Step 5: Plot top 10 bowlers by wickets
top_bowlers = (
    df.groupby('best_bowling')['highest_wickets']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(8,5))
top_bowlers.plot(kind='barh', color='skyblue')
plt.title('Top 10 Bowlers by Total Wickets')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.gca().invert_yaxis()
plt.show()


```
