```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 3

print(x)
plt.plot(x, y)
plt.title("Sonali")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,y)
```

```python
#Object Oriented Way
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 3
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```
