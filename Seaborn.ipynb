```python
df=sns.load_dataset('tips')

df['size'].unique()
```

```python
#dispplot
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
sns.displot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.displot(df['tip'],bins=20,kde=True)
```

```python
#Jointplot
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')

sns.jointplot(x='total_bill', y='tip', data=df,kind='reg')
plt.show()
```

```python
#pair plot
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
#rugplot
sns.rugplot(df['tip'])
```
