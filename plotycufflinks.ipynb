```python
import seaborn as sns
import plotly.express as px

# Load dataset
tips = sns.load_dataset('tips')

# Interactive line plot
fig = px.line(
    tips,
    y='total_bill',
    title='Total Bill Trend',
    labels={'index': 'Index', 'total_bill': 'Total Bill ($)'},
    color_discrete_sequence=['orange']
)

fig.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
avg_tip = tips.groupby('day')['tip'].mean()

avg_tip.plot(kind='bar', color='skyblue')
plt.title('Average Tip by Day')
plt.ylabel('Average Tip')
plt.xlabel('Day')
plt.show()
```

```python
import seaborn as sns
import plotly.express as px

import seaborn as sns
import plotly.express as px
import plotly.io as pio

# ✅ Set renderer for Google Colab
pio.renderers.default = "colab"

# Load dataset
tips = sns.load_dataset('tips')

# Create an interactive plot
fig = px.scatter(
    tips,
    x='total_bill',
    y='tip',
    color='day',
    size='size',
    hover_data=['sex', 'time'],
    title='Total Bill vs Tip Scatter Plot'
)

fig.show()


```

```python
import seaborn as sns
import plotly.express as px
import plotly.io as pio

# Set renderer for Colab
pio.renderers.default = "colab"

# Load dataset
tips = sns.load_dataset('tips')

# Box plot with Plotly Express
fig = px.box(
    tips,
    x='day',
    y='total_bill',
    color='day',
    title='Box Plot of Total Bill by Day'
)

fig.show()
```
